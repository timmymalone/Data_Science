```python
D = np.array([4,14,5,18,10])    # These numbers represent the diamater of an object measured
C = np.array([13,44,16,56,32])  # These numbers represent the circumfrence of an object measured
```

```python
plt.plot(D,C,'r.')
```

```python
x = D #the x values represent the diameter values
y = C #the y valueves equal the circumfrence values
```

```python
m_guess = 0 #the lower the numbers the closer the data is and more accurate your regression line will be
b_guess = 0 #so both of these numbers are here to tell python to run these numbers and give us the one line that best represents the data by giving us a line with the closest number to 0
```

```python
best_loss = 1000000000 #running the system this many times to try different numbers to best the best representation of the data
best_m = 0 #the lower the numbers the closer the data is and more accurate your regression line will be
best_b = 0 

for m_guess in np.linspace(-100,100,100): #seeing what angle the line will be to best match data points
    for b_guess in np.linspace(-100,100,100): # seeing what the y-intercept would be to best start the regression line

        print("Now Testing: ", m_guess, b_guess)
        
        y_guess = m_guess * x + b_guess

        loss = np.sum((y_guess - y)**2)

        if loss < best_loss:
            
            best_loss = loss
            best_m = m_guess
            best_b = b_guess 
```

```python
y_guess = best_m * x + best_b #plotting only the best/closest regression line to the data points.

plt.plot(x,y_guess,'r');# this plots the regression line
plt.plot(D,C,'g.'); #this plots the data points
plt.show()

print(best_loss,best_m,best_b) #this is telling it to print the best line that is the closest to all the data points
```
